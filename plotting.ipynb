```python
# ANT
bc_data = [0.46, 10.3, 12.20, 12.53]
gail_data = [11.4, 13.1, 13.35, 13.81]
irl_data = [0.96, 0.97, 1.1, 1.23]
irl_chi2_data = [0.83, 0.87, 1.43, 1.57]

n_traj = [1, 10, 100, 1000]

datas = [bc_data, gail_data, irl_data, irl_chi2_data]
datas = [np.array(data) for data in datas]

names = ['BC', 'GAIL', 'Chi-X²', 'W1']


bc_std = [0.02, 0.01, 0.003, 0.002]
gail_std = [0.01, 0.008, 0.003, 0.002]
irl_std = [0.02, 0.015, 0.003, 0.002]
wasserstein_std = [0.015, 0.01, 0.005, 0.002]
```

```python
# Example std values (replace with your actual std values)


stds = [np.array(bc_std), np.array(gail_std), np.array(irl_std), np.array(wasserstein_std)]  # Using irl_std for chi2 as well
factor = 40
stds = [std * factor for std in stds]  # Scale stds by a factor of 40
plt.rcParams.update({'font.size': 18})

fig, ax = plt.subplots(figsize=(10, 6))
# Highlight the area between (mean - std) and (mean + std) instead of error bars
for data, std, name in zip(datas, stds, names):
    ax.plot(n_traj, data, label=name, marker='D', linestyle='--', markersize=10)
    ax.fill_between(n_traj, data - std, data + std, alpha=0.2)
ax.set_xlabel('Number of Trajectories')

ax.set_ylabel('Average Reward')
ax.set_title('Sample Efficiency')
ax.set_xscale('log')
ax.set_xticks(n_traj)
ax.legend()
ax.grid()
plt.show()
```
